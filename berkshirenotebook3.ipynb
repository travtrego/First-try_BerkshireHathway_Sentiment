```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA

results = []

for title in df3['Headline']:
    pol_score = SIA().polarity_scores(title) # run analysis
    pol_score['Title'] = title # add headlines for viewing
    results.append(pol_score)
    
results
```

```python
df3['Score'] = pd.DataFrame(results)['compound']
df3
```
